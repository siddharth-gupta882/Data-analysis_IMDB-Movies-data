
# IMDB_Movie_Data:

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('IMDB_Movie_Data.csv')
```

```python
df.head()
```

#### Renaming the columns for the analysis purpose

```python
df = df.rename({'Runtime (Minutes)': 'Runtime_Minutes', 'Revenue (Millions)': 'Revenue_Millions'}, axis=1)
```

```python
df.head()
```

#### Handling the missing value columns with the proper assumtion.

```python
df.isnull().sum()
```

```python
df['Metascore'] = df['Metascore'].fillna(df['Metascore'].mean())
df['Revenue_Millions']=df['Revenue_Millions'].fillna(df['Revenue_Millions'].mean())
```

```python
df['Metascore'].isnull().sum()
df['Revenue_Millions'].isnull().sum()
df.isnull().sum()
```

####  Generating the relationship between each continuous variable 

```python
df.corr()
```

####  Finding the movies directed by 'James Gunn', & 'Ridley Scott' with the rating greater than or equal to 8.

```python
df.loc[(df['Director']=='Ridley Scott')&(df['Rating']>=8)|(df['Director']=='James Gunn')&(df['Rating']>=8),['Title','Director',"Rating"]]
```

####  Displaying all the movies directed by 'Ridley Scott'.

```python
df.loc[df['Director']=='Ridley Scott',['Title']]
```

####  Displaying all the director's who got the rating >= 8 for their movies.

```python
df.loc[df['Rating']>=8,['Title','Director']]
```

####  all the movies directed by 'James Gunn', & 'Ridley Scott'.

```python
df.loc[(df['Director']=='Ridley Scott')|(df['Director']=='James Gunn'),['Title','Director']]
```

#### all the movies with the ratings between 8 & 9 and also the films with the highest rating.

```python
df.loc[((df['Rating'] >=8) & (df['Rating'] <= 9)) ,['Title']]
```

####  Find the movies directed by both 'Christopher Nolan' and 'Ridley Scott'

```python
df.loc[(df['Director']=='Ridley Scott')|(df['Director']=='Christopher Nolan'),['Title','Director']]
```

####   all movies that were released between 2006 and 2010, have a rating above 8.0, but made below the 50th percentile in revenue.

```python
df[((df['Year'] >= 2006) & (df['Year'] <= 2010))
      & (df['Rating'] > 8.0)
      & (df['Revenue_Millions'] < df['Revenue_Millions'].quantile(0.50))]
```

####  all movies that were released between 2006 and 2010, have a rating above 8.0, but made below the 25th percentile in revenue.

```python
df[((df['Year'] >= 2006) & (df['Year'] <= 2010))
      & (df['Rating'] > 8.0)
      & (df['Revenue_Millions'] < df['Revenue_Millions'].quantile(0.25))]
```

#### Feedback for all movies with 'Good' if the movie 'rating' is >= 8 and 'Average' if the movie rating doesn't follow the condition given.

```python
def Feedback(rating):
    if rating >=8:
        return 'Good'
    else:
        return 'Average'
```

```python
df['Feedback'] = df['Rating'].apply(Feedback)
df[['Title','Director','Rating','Feedback']].head()
```
